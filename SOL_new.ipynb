```python
import matplotlib.pyplot as plt
v_data = r_file
v_data.head(5)
```

```python
count_data =vis['bedrooms'].value_counts()
# count_data.rename({'1': '1 bedroom', '2': '2 bedrooms', '3': '3 bedrooms','4':'4 bedrooms','5': '5 bedrooms',
#                         '6': '6 bedrooms','10':'10 bedrooms','7': '7 bedrooms'}, inplace=True)
count_data
```

```python
# rename the values of the index 
cou= count_data.rename({'1': '1 bedroom', '2': '2 bedrooms', '3': '3 bedrooms','4':'4 bedrooms','5': '5 bedrooms',
                        '6': '6 bedrooms','10':'10 bedrooms','7': '7 bedrooms'}, inplace=True)
cou
```

```python
# # plotting pie chart
labels = ["","2 bedrooms","","4 bedrooms","5 bedrooms", "6 bedrooms", "7 bedrooms", "10 bedrooms"]
plt.style.use("default")
plot = plt.pie(count_data,autopct = "%1.1f%%",labeldistance=None, startangle=0)
plt.title("Proportion of bedrooms in Airbnb listing")
# set legend outside the pie
plt.tight_layout()
percents = count_data.to_numpy() * 100 / count_data.to_numpy().sum()
plt.legend( bbox_to_anchor=(1.35,1.1), loc='upper right',
            labels=['%s, %1.1f %%' % (l, s) for l, s in zip(count_data.index,percents)])
plt.show()
```
